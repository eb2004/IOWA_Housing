```python
import numpy as np
import pandas as pd
import os
import tarfile
import io
import matplotlib
import matplotlib.pyplot as plt
from pandas.plotting import scatter_matrix
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import confusion_matrix, precision_score, recall_score, f1_score
from sklearn.preprocessing import OrdinalEncoder
from sklearn.preprocessing import LabelEncoder
```

```python
train = pd.read_csv('train.csv')
test = pd.read_csv('test.csv')
```

```python
train = train.drop(columns=['LotFrontage','Alley', 'FireplaceQu', 'PoolQC', 'Fence', 'MiscFeature', 'HouseStyle'])
test = test.drop(columns=['LotFrontage','Alley', 'FireplaceQu', 'PoolQC', 'Fence', 'MiscFeature', 'HouseStyle'])
```

```python
train = train.dropna(subset=['MasVnrType', 'Electrical', 'BsmtFinType2', 'BsmtExposure', 'GarageCond'])
```

```python
train = train.drop(columns=['Id'])
test = test.drop(columns=['Id'])
```

```python
encoder = LabelEncoder()
train['MSZoning'] = encoder.fit_transform(train['MSZoning'])
test['MSZoning'] = encoder.fit_transform(test['MSZoning'])
```

```python
train['Street'] = encoder.fit_transform(train['Street'])
test['Street'] = encoder.fit_transform(test['Street'])
```

```python
train['LotShape'] = encoder.fit_transform(train['LotShape'])
test['LotShape'] = encoder.transform(test['LotShape'])

train['LandContour'] = encoder.fit_transform(train['LandContour'])
test['LandContour'] = encoder.transform(test['LandContour'])

train['Utilities'] = encoder.fit_transform(train['Utilities'])
test['Utilities'].fillna(value='AllPub', inplace=True)
test['Utilities'] = encoder.transform(test['Utilities'])

train['LotConfig'] = encoder.fit_transform(train['LotConfig'])
test['LotConfig'] = encoder.transform(test['LotConfig'])

train['LandSlope'] = encoder.fit_transform(train['LandSlope'])
test['LandSlope'] = encoder.transform(test['LandSlope'])

train['Neighborhood'] = encoder.fit_transform(train['Neighborhood'])
test['Neighborhood'] = encoder.transform(test['Neighborhood'])

train['Condition1'] = encoder.fit_transform(train['Condition1'])
test['Condition1'] = encoder.transform(test['Condition1'])

train['Condition2'] = encoder.fit_transform(train['Condition2'])
test['Condition2'] = encoder.transform(test['Condition2'])

train['BldgType'] = encoder.fit_transform(train['BldgType'])
test['BldgType'] = encoder.transform(test['BldgType'])

train['RoofStyle'] = encoder.fit_transform(train['RoofStyle'])
test['RoofStyle'] = encoder.transform(test['RoofStyle'])

train['RoofMatl'] = encoder.fit_transform(train['RoofMatl'])
test['RoofMatl'] = encoder.transform(test['RoofMatl'])

train['Exterior1st'] = encoder.fit_transform(train['Exterior1st'])
test['Exterior1st'].fillna(value='VinylSd', inplace=True)
test['Exterior1st'] = encoder.transform(test['Exterior1st'])

train['Exterior2nd'] = encoder.fit_transform(train['Exterior2nd'])
test['Exterior2nd'] = encoder.transform(test['Exterior2nd'])

train['MasVnrType'] = encoder.fit_transform(train['MasVnrType'])
test['MasVnrType'] = encoder.transform(test['MasVnrType'])

train['ExterQual'] = encoder.fit_transform(train['ExterQual'])
test['ExterQual'] = encoder.transform(test['ExterQual'])

train['ExterCond'] = encoder.fit_transform(train['ExterCond'])
test['ExterCond'] = encoder.transform(test['ExterCond'])

train['Foundation'] = encoder.fit_transform(train['Foundation'])
test['Foundation'] = encoder.transform(test['Foundation'])

train['BsmtQual'] = encoder.fit_transform(train['BsmtQual'])
test['BsmtQual'] = encoder.transform(test['BsmtQual'])

train['BsmtCond'] = encoder.fit_transform(train['BsmtCond'])
test['BsmtCond'] = encoder.transform(test['BsmtCond'])

train['BsmtExposure'] = encoder.fit_transform(train['BsmtExposure'])
test['BsmtExposure'] = encoder.transform(test['BsmtExposure'])

train['BsmtFinType1'] = encoder.fit_transform(train['BsmtFinType1'])
test['BsmtFinType1'] = encoder.transform(test['BsmtFinType1'])

train['BsmtFinType2'] = encoder.fit_transform(train['BsmtFinType2'])
test['BsmtFinType2'] = encoder.transform(test['BsmtFinType2'])

train['Heating'] = encoder.fit_transform(train['Heating'])
test['Heating'] = encoder.transform(test['Heating'])

train['CentralAir'] = encoder.fit_transform(train['CentralAir'])
test['CentralAir'] = encoder.transform(test['CentralAir'])

train['Electrical'] = encoder.fit_transform(train['Electrical'])
test['Electrical'] = encoder.transform(test['Electrical'])

train['KitchenQual'] = encoder.fit_transform(train['KitchenQual'])
test['KitchenQual'] = encoder.transform(test['KitchenQual'])

train['Functional'] = encoder.fit_transform(train['Functional'])
test['Functional'] = encoder.transform(test['Functional'])

train['GarageType'] = encoder.fit_transform(train['GarageType'])
test['GarageType'] = encoder.transform(test['GarageType'])

train['GarageFinish'] = encoder.fit_transform(train['GarageFinish'])
test['GarageFinish'] = encoder.transform(test['GarageFinish'])

train['GarageQual'] = encoder.fit_transform(train['GarageQual'])
test['GarageQual'] = encoder.transform(test['GarageQual'])

train['GarageCond'] = encoder.fit_transform(train['GarageCond'])
test['GarageCond'] = encoder.transform(test['GarageCond'])

train['PavedDrive'] = encoder.fit_transform(train['PavedDrive'])
test['PavedDrive'] = encoder.transform(test['PavedDrive'])

train['SaleType'] = encoder.fit_transform(train['SaleType'])
test['SaleType'] = encoder.transform(test['SaleType'])

train['SaleCondition'] = encoder.fit_transform(train['SaleCondition'])
test['SaleCondition'] = encoder.transform(test['SaleCondition'])

train['HeatingQC'] = encoder.fit_transform(train['HeatingQC'])
test['HeatingQC'] = encoder.transform(test['HeatingQC'])
```

```python
X, y = train.drop(columns=['SalePrice']), train['SalePrice']
pd.set_option("display.max_rows", 5, "display.max_columns", None)
print(X)
```

```python
from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2)
```

```python
from sklearn.decomposition import PCA

pca = PCA(n_components=10)
X_train_reduced = pca.fit_transform(X_train)
X_test_reduced = pca.transform(X_test)
```

```python
X_train.shape
```

```python
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error

lin_reg = LinearRegression()

lin_reg.fit(X_train, y_train)
lin_reg_predictions = lin_reg.predict(X_test)
lin_reg_mse = mean_squared_error(lin_reg_predictions, y_test)
```

```python
print(lin_reg_mse)
```

```python
lin_reg2 = LinearRegression()

lin_reg2.fit(X_train_reduced, y_train)
lin_reg2_predictions = lin_reg2.predict(X_test_reduced)
lin_reg2_mse = mean_squared_error(lin_reg2_predictions, y_test)

print(lin_reg2_mse)
```

```python
from sklearn.preprocessing import PolynomialFeatures
from sklearn.linear_model import SGDRegressor

for i in range(1,3):
    poly_features = PolynomialFeatures(degree=i)
    X_train_poly = poly_features.fit_transform(X_train)
    X_test_poly = poly_features.fit_transform(X_test)

    lin_regressor = LinearRegression()
    lin_regressor.fit(X_train_poly, y_train)
    lin_regressor_pred = lin_regressor.predict(X_test_poly)
    lin_regressor_mse = mean_squared_error(lin_regressor_pred, y_test)
    print(i, np.sqrt(lin_regressor_mse))
```
